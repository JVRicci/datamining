```python
def retornar_str(coluna):
    if len(coluna) < 20:
        return coluna.title()
    else:
        return coluna.lower()    
```

```python
df[['NO_ENTIDADE']].applymap(lambda x: retornar_str(x)) 
```

```python
def retorna_descricao_situacao(codigo):
    if codigo == 1:
        return 'Em Atividade'
    elif codigo == 2:
        return 'Paralisada'
    elif codigo == 3:
        return 'Extinta (ano do Censo)'
    elif codigo == 4:
        return 'Extinta em Anos Anteriores'
    else:
        return 'Não conhecido'
```

```python
df['TP_SITUACAO_FUNCIONAMENTO_DESC'] = df[['TP_SITUACAO_FUNCIONAMENTO']].applymap(retorna_descricao_situacao)
```
